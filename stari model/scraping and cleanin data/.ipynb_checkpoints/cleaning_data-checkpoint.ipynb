```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from tqdm import tqdm, tqdm_notebook

```

```python
df = pd.read_csv('stanovi_data.csv')
df.columns = ['url', 'naziv', 'cijena', 'lokacija', 'lat_lng' ]
```

```python
df.head()
```

```python
df.info()
```

```python
df['lat_lng'][0]
```

```python
# za latitude i longitude
coord = []
def cleaning(strings):
    for string in tqdm_notebook(strings):
        try:        
            x, y = (string.split('=')[1]).split(',')
            x = float(x)
            y = y.split('&')[0]
            coord_turn = (x,y)
            coord.append(coord_turn)
        except:
            coord_turn = ('NaN', 'NaN')
            coord.append(coord_turn)
```

```python
extra = pd.DataFrame(df['lat_lng'])
extra.apply(cleaning)
```

```python
df1 = pd.DataFrame(coord, columns=['lat', 'lng'])
df1.head()
```

```python
df = df.join(df1, lsuffix='_df', rsuffix='_df1')
```

```python
df.head()
```

```python
df = df.drop('lat_lng', axis=1)
```

```python
df.head()
```

```python
kolone = ['Kvadrata',
 'Broj soba','Adresa',
 'Godina izgradnje','Primarna orjentacija']
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
linkovi = df['url']
linkovi = pd.DataFrame(linkovi)
linkovi.head()
```

```python
# za dodatne podatke o stanovima(kvadratura, opca lokacija, adresa)

col = []
val = []
for i in tqdm_notebook(range(len(linkovi))):
    val_turn = []
    resp = requests.get(linkovi['url'][i])
    if str(resp) == '<Response [404]>':
        val.append('NaN')
    
    soup = BeautifulSoup(resp.text, 'html.parser')
    try:
        section1 = soup.find('div', id='dodatnapolja1')

        sections = section1.select('.df')
        for section in sections:
            x = section.select('.df1')[0].text.strip()
            y = section.select('.df2')[0].text.strip()
            
            if x in kolone:
                col.append(x)
                if y=='':
                    y='NaN'
                val_turn.append(y)            
                  
                
    except:
        val.append('NaN')
        continue
    
    val.append(val_turn)

```

```python
val2 = []
```

```python
for i in val:
    if i == 'NaN':
        val2.append(['NaN'])
    else:
        val2.append(i)
```

```python
val2 = pd.DataFrame(val2, columns=kolone)
```

```python
val2
```

```python
df = df.join(val2, lsuffix='_df', rsuffix='_df2')
```
