```python
# za latitude i longitude
coord = []
def cleaning(strings):
    for string in tqdm_notebook(strings):
        try:        
            x, y = (string.split('=')[1]).split(',')
            x = float(x)
            y = y.split('&')[0]
            coord_turn = (x,y)
            coord.append(coord_turn)
        except:
            coord_turn = ('NaN', 'NaN')
            coord.append(coord_turn)
```

```python
extra = pd.DataFrame(df['lat_lng'])
extra.apply(cleaning)
```

```python
df1 = pd.DataFrame(coord, columns=['lat', 'lng'])
df1.head()
```

```python
df = df.join(df1, lsuffix='_df', rsuffix='_df1')
```
